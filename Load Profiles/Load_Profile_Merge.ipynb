```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
# Path to the folder containing the CSV files
folder_path = "."  # Replace with your folder path


# List all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith(".csv")]

# Get the list of CSV files and sort them numerically
csv_files = sorted(
    [file for file in os.listdir(folder_path) if file.endswith(".csv")],
    key=lambda x: int(''.join(filter(str.isdigit, x)))  # Extract numeric part for sorting
)

# List to store processed DataFrames
processed_dataframes = []

# Process each file
for file in csv_files:
    # Load the CSV file
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    
    # Fix invalid time format (replace 24:00:00 with 00:00:00)
    df['time'] = df['time'].replace('24:00:00', '00:00:00')
    
    # Ensure the time column is in datetime format
    df['time'] = pd.to_datetime(df['time'], format='%H:%M:%S')  # Keep only time
    
    # Set the time column as index
    df.set_index('time', inplace=True)
    
    # Resample to 30-minute intervals and calculate the mean
    resampled_df = df.resample('30T').mean()
    
    # Round to 3 decimals
    resampled_df = resampled_df.round(3)
    
    # Rename column to indicate the file it came from
    resampled_df.rename(columns={"mult": file.replace(".csv", "")}, inplace=True)
    
    # Store the resampled DataFrame
    processed_dataframes.append(resampled_df)

# Concatenate all processed DataFrames into a single DataFrame
merged_df = pd.concat(processed_dataframes, axis=1)
merged_df.index = merged_df.index.time

# Save the merged DataFrame as an .npy file
output_path = os.path.join(folder_path, "merged_load_profiles.npy")
np.save(output_path, merged_df.to_numpy())

# Save the merged DataFrame as a CSV for easier inspection (optional)
merged_df.to_csv(os.path.join(folder_path, "merged_load_profiles.csv"))

# Print the output path for the saved .npy file
print(f"Data saved to: {output_path}")
```
