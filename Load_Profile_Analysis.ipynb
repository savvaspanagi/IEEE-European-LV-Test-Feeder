```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Load_Profiles = pd.read_csv("Load_Profiles.csv", index_col=0)
```

```python
# Plot all load profiles
plt.figure(figsize=(12, 6))  # Set the figure size
# for column in Load_Profiles.columns:
plt.plot(Load_Profiles.sum(axis=1))

# Add labels, legend, and title
plt.xlabel("Time")
plt.ylabel("multiplier")
plt.title("Load Profiles")
# plt.legend(loc="best", fontsize="small")  # Adjust the legend position
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
